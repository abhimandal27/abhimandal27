```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df= pd.read_csv('/content/WorldCupMatches.csv')
df
```

```python
#  Number of rows and columns
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

```python
# name of each of the columns
df.columns
```

```python
# Print specific range using slicing (e.g., first 5 rows)
print("\nFirst 5 rows of the dataset:")
df.head()
```

```python
# Summarization statistics using Python libraries
print("\nSummary statistics:")
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:\n", missing_values)
```

```python
# Replace missing numerical values with mean or median
for column in df.select_dtypes(include=['number']):
  if df[column].isnull().any():
    if df[column].skew() > 1:  # Check for skewness
      df[column].fillna(df[column].median(), inplace=True)  # Use median for skewed data
      print(f"Missing values in '{column}' replaced with median.")
    else:
      df[column].fillna(df[column].mean(), inplace=True)  # Use mean for non-skewed data
      print(f"Missing values in '{column}' replaced with mean.")
```

```python
# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=df)
plt.show()
```

```python
# v) Distribution plot
plt.figure(figsize=(8, 6))
sns.histplot(df['Home Team Goals'], kde=True)
plt.show()
```

```python
# iv) Barplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Home Team Name', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# i) Pairplot
sns.pairplot(df)
```

```python
# ii) Boxplot
sns.boxplot(data=df)
```

```python
# iii) Heatmap
# Convert date columns to datetime objects if needed
for col in df.columns:
    if df[col].dtype == 'object':  # Check if column is of object type (potentially string)
        try:
            df[col] = pd.to_datetime(df[col])  # Attempt to convert to datetime

        except:
            pass  # Skip if conversion fails

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number', 'datetime'])  # Include datetime columns
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
```
